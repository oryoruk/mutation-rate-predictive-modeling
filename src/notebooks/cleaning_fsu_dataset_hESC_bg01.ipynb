```python
def parse_FSU_rt_dataset(file_name, end_inclusive = False, with_win_size = False, with_chrom_mid = True, chrom='all'):
    #to-do: skiprows is variable (16-17), change this next line
    FSU_rt_dataset = pd.read_csv(file_name,delimiter = '\t',header=None, names = ['ID','Gene_Name','chrom','chrom_start','chrom_end','rt_score'] , skiprows=16)
    if end_inclusive: 
        FSU_rt_dataset.chrom_end = FSU_rt_dataset.chrom_end + 1
    FSU_rt_dataset["win_size"] = FSU_rt_dataset.chrom_end - FSU_rt_dataset.chrom_start
    if with_chrom_mid:
        FSU_rt_dataset["chrom_mid"] = FSU_rt_dataset.chrom_start +  (FSU_rt_dataset.win_size) /int(2)
        FSU_rt_dataset.chrom_mid = FSU_rt_dataset.chrom_mid.astype(int)
    if not with_chrom_mid: FSU_rt_dataset.drop('win_size',axis=1,inplace=True)
    FSU_rt_dataset.drop('ID',axis=1,inplace=True)
    FSU_rt_dataset.drop('Gene_Name',axis=1,inplace=True)
    FSU_rt_dataset.chrom = FSU_rt_dataset.chrom.str.strip('chr')
    if chrom == "all": return FSU_rt_dataset
    return FSU_rt_dataset[FSU_rt_dataset.chrom==chrom]
```

```python
rt_bg01_avg = parse_FSU_rt_dataset(hESC_dir+'RT_BG01_ESC_All_avg.txt', end_inclusive=True, with_win_size=True)
```
